```python
def compute_cost(y,y_pred):
  n = len(y)
  cost = (1/(2*n))*np.sum((y-y_pred)**2)
  return cost
```

```python
cost = compute_cost(Y,y_pred)
print("intial cost:", cost)
```

```python
m=0.0
c=0.0
y_pred =predict(X_norm,n,c)
cost = compute_cost(Y,y_pred)
print("intial cost:", cost)
```

```python
learning_rate=0.01
epochs =1000
print("learning rate",learning_rate)
print("epochs",epochs)
```

```python
n=len(y)
n
```

```python
cost_History=[]
```

```python
for epoch in range(epochs):
  #step1
  y_pred = predict(X_norm,m,c)
  #step2
  dn = (1/n)*np.sum((y_pred-Y)*X_norm)
  dc = (1/n)*np.sum(y_pred-Y)
  #step3
  n_m = m-learning_rate*dn
  n_c = c-learning_rate*dc
  #step4
  m=n_m
  c=n_c
  #step5
  cost = compute_cost(Y,y_pred)
  cost_History.append(cost)
  if epoch%100==0:
    print("epoch is",epoch,"cost is",cost)
```

```python
final_slope=n
final_intercept=c
print("final slope",final_slope)
print("final intercept",final_intercept)
```

```python
def predict_salary(YearsExperience,m,c,x_min,x_max):
  x=np.array(YearsExperience)
  x_n=(x-x_min)/(x_max-x_min)
  sal_pred=m*x_n+c
  return sal_pred
```

```python
years=5
predicted_salary=predict_salary(
    years,
    final_slope,
    final_intercept,
    X.min(),
    X.max())
print(f"predicted salary for {years} years of experience")
print(f"predicted salary is {predicted_salary}")
```
